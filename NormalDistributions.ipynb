# The Normal Distribution

Imagine that you are sampling from a small number (200) cats

Imagine that the mean weight of cats is 12 lbs with a standard deviation of 3 lbs

```python
mu = 12
sigma = 3 
```

```python
n = 200
```

```python
sample = pd.Series(np.random.normal(mu, sigma, n))
```

```python
#sample #remove hash to show
```

```python
sample.plot.hist(grid=False, bins=5, rwidth=0.9,
                   color='#607c8e')
plt.title('Weight for Cats')
plt.ylabel('Counts')
plt.xlabel('Weight(lbs)')
plt.grid(axis='y', alpha=0.75)
```

Now suppose that we increase the number of cats we sample... what happens to the distribution?

### Let Y be the random variable for the weight of cats
If the Y is normally distributed, then the value for the weight of cats can be found using:

$$Y = \frac{1}{{\sigma \sqrt {2\pi } }}e^{{{ - \left( {x - \mu } \right)^2 } \mathord{\left/ {\vphantom {{ - \left( {x - \mu } \right)^2 } {2\sigma ^2 }}} \right.} {2\sigma ^2 }}}$$


* The expected value of Y is &mu;
* The standard deviation of Y is &sigma;

We say that $Y \sim N(0,\sigma)$

```python
import scipy.stats as stats
import math

mu = 12
sigma = 3
y = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(y, stats.norm.pdf(y, mu, sigma))
plt.title('Weight for Cats')
plt.ylabel('Probability Density')
plt.xlabel('Weight(lbs)')
plt.grid(axis='y', alpha=0.75)
```

## Showing the probability of a range:

```python
from scipy.stats import norm
def draw_plot(x, cond, mu, sigma):
    y = norm.pdf(x, mu, sigma)
    z = x[cond]
    plt.plot(x, y)
    plt.fill_between(z, 0, norm.pdf(z, mu, sigma))
    plt.show()
x = np.arange(2.5,21)
x1= 11
draw_plot(x, x<x1, 12, 3)
```

```python
norm(mu,sigma).pdf(x1)
```

```python
for i in range(7,17):
    x1 = i
    draw_plot(x, x<i, 12, 3)   
```

## Standard Normal Distribution
The standard normal distribution is the distribution of standard deviations away from the mean. In other words:

$$Z=\frac{Y-\mu}{\sigma}$$

The random variable Z is normally distributed with a mean of 0 and standard deviation of 1.

$$Z \sim N(0,1)$$

We can easily convert between Y and Z:
* Y = 10 
$$Z = \frac{10-12}{3}=-0.67$$
* Z = 5
$$5 =\frac{Y-12}{3} \rightarrow Y=5(3)+12=27$$

### The 68-95-99.8 Rule
If Y is normally distributed with mean $\mu$ and standard deviation $\sigma$:
1. $P(\mu-\sigma < x< \mu +\sigma) = 0.68$
2. $P(\mu-2\sigma < x< \mu +2\sigma) = 0.95$
3. $P(\mu-3\sigma < x< \mu +3\sigma) = 0.998$
